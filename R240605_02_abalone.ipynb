#### 껍질무게와 나이테

```python
var ='Shell weight'
plt.scatter(x=df[var],y=df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Shell weight']>0.6) & (df['Rings']<25)].index, inplace=True)
df.drop(df[(df['Shell weight']<1.0) & (df['Rings']>25)].index, inplace=True)
```

```python
var ='Shell weight'
plt.scatter(x=df[var],y=df['Rings'])
plt.grid(True)
```

```python
var ='Viscera weight'
plt.scatter(x=df[var],y=df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Viscera weight']>0.5) & (df['Rings']<20)].index, inplace=True)
df.drop(df[(df['Viscera weight']<0.5) & (df['Rings']>=25)].index, inplace=True)
var ='Viscera weight'
plt.scatter(x=df[var],y=df['Rings'])
plt.grid(True)
```

```python
# 길이 이상치 제거
var ='Length'
plt.scatter(x=df[var],y=df['Rings'])
plt.grid(True)
```

```python
df.drop(df[(df['Length']<0.1) & (df['Rings']<2)].index, inplace=True)
# df.drop(df[(df['Viscera weight']<0.5) & (df['Rings']>=25)].index, inplace=True)
var ='Length'
plt.scatter(x=df[var],y=df['Rings'])
plt.grid(True)
```

```python
# 순살 무게 이상치 제거

var = 'Shucked weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
df.drop(df[(df['Shucked weight']>= 1) & (df['Rings'] < 20)].index, inplace=True)
df.drop(df[(df['Shucked weight']<1) & (df['Rings'] > 20)].index, inplace=True)

var = 'Shucked weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
# 전체 무게 이상치 제거
var = 'Whole weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
df.drop(df[(df['Whole weight']> 2.0) & (df['Rings'] <= 20)].index, inplace=True)

var = 'Whole weight'
plt.scatter(x = df[var], y = df['Rings'])
plt.grid(True)
```

```python
# 직경 이상치 제거
var = 'Diameter'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
df.drop(df[(df['Diameter']<0.1) & (df['Rings']< 2.5)].index, inplace=True)

var = 'Diameter'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
```

```python
# 높이(두께) 이상치 제거
var = 'Height'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
df.drop(df[(df['Height']>0.4) & (df['Rings'] < 15)].index, inplace=True)

var = 'Height'
plt.scatter(x = df[var], y =df['Rings'])
plt.grid(True)
```
